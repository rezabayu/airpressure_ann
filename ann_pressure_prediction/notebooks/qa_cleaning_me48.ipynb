```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns

RAW  = "/home/rzby/airpressure_ann/ann_pressure_prediction/data/interim/three_h-97430.parquet"
CLEAN= "/home/rzby/airpressure_ann/ann_pressure_prediction/data/processed/clean_97430.parquet"

# Before cleaning
raw = pd.read_parquet(RAW)   # adjust if needed
miss_raw = raw.isna().sum()

# After cleaning
clean = pd.read_parquet(CLEAN)
miss_clean = clean.isna().sum()

# --- Visual proof ---
plt.figure(figsize=(12,4))
plt.bar(miss_raw.index, miss_raw.values)
plt.xticks(rotation=90); plt.ylabel("Missing values"); plt.title("Sultan Babullah Station | Before cleaning")
plt.tight_layout(); plt.show()

plt.figure(figsize=(12,4))
plt.bar(miss_clean.index, miss_clean.values)
plt.xticks(rotation=90); plt.ylabel("Missing values"); plt.title("Sultan Babullah Station | After cleaning")
plt.tight_layout(); plt.show()

assert miss_clean.sum() == 0, "NaNs still present!"
print("✅  No missing data after cleaning.")
```
